# Задание 1.4.3 Анализ зависимостей

```python
age = []
sex = []
bmi = []
children = []
smoke = []
region = []
charges = []
```

```python
with open ("data.txt", mode = "r") as file:
  for human in file.readlines():
    age.append(int(human.split(',')[0]))
    if human.split(",")[1] == "male":
      sex.append(1)
    else:
      sex.append(0)
    bmi.append(float(human.split(',')[2]))
    children.append(int(human.split(',')[3]))
    if human.split(",")[4] == "yes":
      smoke.append(0)
    else:
      smoke.append(1)
    if human.split(",")[5] == "northeast":
      region.append(0)
    elif human.split(",")[5] == "southeast":
      region.append(1)
    elif human.split(",")[5] == "northwest":
      region.append(2)
    else:
      region.append(3)
    charges.append(float(human.split(',')[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoke = np.array(smoke)
region = np.array(region)
charges = np.array(charges)
```

```python
plt.scatter(age, charges)
plt.show()
```

На графике выше очевидно изменение стоимости страховки с возрастом.

```python
plt.scatter(sex, charges)
plt.show()
```

На графике выше представлена зависимость стоимости страховки и пола. Стоимость страховки для мужчин может быть больше. Возможно, предположить, что это связано с более низкой прододжительностью жизни мужчин.

```python
plt.scatter(bmi, charges)
plt.show()
```

На графике выше представлена зависимость индекса массы тела и стоимости страховки. Прямой зависимости не выявлено. Однако, следует выделить два кластера значений (Люди с ИМС > 30 чаще платят за страховку больше).

```python
plt.scatter(smoke, charges)
plt.show()
```

Некурящие платят за страховку меньше курящих. Различия статистически значимы.

```python
plt.scatter(children, charges)
plt.show()
```

Люди имеющие 4 и более детей платят за страховку меньше.
Возможно, здесь имеет место факторы о которых мы не знаем, например льгота на страховку для многгодетных родителей.

```python
plt.scatter(region, charges)
plt.show()
```

Стоимость по регионам распределна относительно равномерно, следует выделит отдельные выбросы, характерные для южных регионов.
